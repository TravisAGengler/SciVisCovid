### Cartopy example

```python
fig = plt.figure(figsize=(16,8), dpi=90)
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([-180,180,-90,90])
ax.coastlines()

ax.plot(-117.1625, 32.715, 'ro', markersize=1, transform=ccrs.Geodetic(), alpha=0.5)
ax.plot( -118.24368,34.05223, 'ro', markersize=1, transform=ccrs.Geodetic(), alpha=0.5)
# fig.savefig(f"{OUTPUT_PATH}/test_image.png", format='png')

plt.show()
```
